# Problem Statement:

**Dream Housing Finance company deals in all home loans. They have a presence across all urban, semi-urban, and rural areas. Customer-first applies for a home loan after that company validates the customer eligibility for a loan.** 

*The company wants to automate the loan eligibility process (real-time) based on customer detail provided while filling the online application form. These details are Gender, Marital Status, Education, Number of Dependents, Income, Loan Amount, Credit History, and others. To automate this process, they have given a problem to identify the customer's segments, those are eligible for loan amount so that they can specifically target these customers.*



### Data source : From Kaggle 
<a href="https://www.kaggle.com/datasets/vikasukani/loan-eligible-dataset/data">Data Set link</a>

## Data Definition:

Input variables:

1) **Loan_ID :** Unique Loan ID

2) **Gender :** Male/ Female

3) **Married :** Applicant married (Y/N)

4) **Dependents :** Number of dependents

5) **Education :** Applicant Education (Graduate/ Under Graduate)

6) **ApplicantIncome :** Applicant Income

7) **CoapplicantIncome :** Coapplicant Income

8) **LoanAmount :** Loan amount in thousands

9) **Loan_Amount_Term	 :** Term of a loan in months

10) **Credit_History :** credit history meets guidelines

11) **Property_Area :** Urban/ Semi-Urban/ Rural

12) **Loan_Status :** Loan approved (Y/N)

# Load The Loan Data Set

```python
# Missing values 

def missing_data(data):
    
    total = data.isnull().sum().sort_values(ascending = False)
    
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    
    return pd.concat([total,percent],axis=1,keys=['Total','Percentile'])

missing_data(loan_data)
```

```python
# Droping the missig values
loan_data_1 = loan_data.dropna()
loan_data_1
```

## Replacing ojbects columns into numeric

```python
# Replace values in the 'Gender' column

loan_data_1['Gender'].replace({'Male': 0, 'Female': 1}, inplace=True)
```

```python
# Replace values in the 'Married' column

loan_data_1['Married'].replace({'No': 0, 'Yes': 1}, inplace=True)
```

```python
# Replace values in the 'Education' column

loan_data_1['Education'].replace({'Not Graduate': 0, 'Graduate': 1}, inplace=True)
```

```python
# Replace values in the 'Self_Employed' column

loan_data_1['Self_Employed'].replace({'No': 0, 'Yes': 1}, inplace=True)
```

```python
# Replace values in the 'Loan_Status' column

loan_data_1['Loan_Status'].replace({'N': 0, 'Y': 1}, inplace=True)
```

```python
# Removing  the un neccesorry columns

loan_data_2 = loan_data_1.drop(['Loan_ID','Dependents','Property_Area'],axis=1)
```

```python
# Converting loan ammount and term 
loan_data_2['LoanAmount'] *= 1000
loan_data_2['Loan_Amount_Term']/= 12
```

```python
loan_data_2 
```

```python
# Splitting the dependent and independent variables

X = loan_data_2.drop('Loan_Status',axis=1)
y = loan_data_2[['Loan_Status']]
```

## Split the data into training and test sets

```python
# split data into train subset and test subset for predictor and target variables

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

print("Shape of is X_train :",X_train.shape)
print("Shape of is X_test :",X_test.shape)
print("Shape of is y_train :",y_train.shape)
print("Shape of is y_test :",y_test.shape)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

## LogisticRegression

```python
# Importing the LogisticRegression from sklearn

from  sklearn.linear_model import LogisticRegression

#intantiate the LogisticRegression

Logistic_Regression = LogisticRegression(n_jobs = 15,C=10,random_state=1)

# fit the LogisticRegression with training dataset

Logistic_Regression_model = Logistic_Regression.fit(X_train, y_train)
```

```python
# predict the values on test dataset using predict()

y_pred = Logistic_Regression_model.predict(X_test)
```

```python
# predict  probality of the values on test dataset using predict_proba()

y_pred_prob = Logistic_Regression_model.predict_proba(X_test)[:,1]
```

```python
# create a generalized function to calculate the metrics values for test set
def get_test_report(model):
    
    # return the performace measures on test set
    return(classification_report(y_test, y_pred))
```

```python
# compute the performance measures on test data
# call the function 'get_test_report'
# pass the logstic regression model to the function

test_report = get_test_report(Logistic_Regression_model)

# print the performace measures

print(test_report)
```

## RandomForestClassifier

```python
# Importing the RandomForestClassifier from sklearn

from sklearn.ensemble import RandomForestClassifier

#intantiate the RandomForestClassifier

RF_model = RandomForestClassifier(n_estimators = 500, random_state = 1, verbose = 1, n_jobs = -1)

# fit the RandomForestClassifier with training dataset

RF_model = RF_model.fit(X_train,y_train)
```

```python
# predict the values on test dataset using predict()

y_pred = RF_model.predict(X_test)
```

```python
# predict  probality of the values on test dataset using predict_proba()

y_pred_prob = RF_model.predict_proba(X_test)[:,1]
```

```python
# compute the performance measures on test data
# call the function 'get_test_report'
# pass the RandomForestClassifier model to the function

test_report = get_test_report(RF_model)

# print the performace measures

print(test_report)
```
